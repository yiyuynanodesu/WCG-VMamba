```python
import pandas as pd
import os
import shutil

# 读取csv文件
path = '../dataset/TomatoPestAndDiseaseDataset/valid/'
df = pd.read_csv(path + '_classes.csv')

# 遍历每一行
for index, row in df.iterrows():
    # 遍历除了第一列以外的每一列
    for col in df.columns[1:]:
        # 如果该位置是1
        if row[col] == 1:
            # 获取该列名
            folder_name = col
            # 创建文件夹
            if not os.path.exists(path + folder_name):
                os.makedirs(path + folder_name)
            # 将该行第一列的文件存入该文件夹中
            if not os.path.exists(path + folder_name + '/' + row[df.columns[0]]):
                shutil.move(path + row[df.columns[0]], path + folder_name + '/' + row[df.columns[0]])
```
